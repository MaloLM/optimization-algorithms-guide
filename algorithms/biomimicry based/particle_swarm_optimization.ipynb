# Particle swarm optimisation

## Particle swarm : Qu'est-ce que c'est ?

L'algorithme d'optimisation par essaims de particules (Particle Swarm Optimization - PSO) est une technique d'optimisation inspirée par le comportement social des oiseaux et des bancs de poissons. Cet algorithme appartient à la catégorie des algorithmes d'optimisation par essaim (swarm intelligence), qui sont des méthodes où de multiples solutions potentielles coopèrent et concourent entre elles pour trouver la meilleure solution à un problème donné.

Voici comment fonctionne le PSO en termes généraux :

1. **Initialisation :** L'algorithme commence par créer un essaim de particules, où chaque particule représente une solution potentielle au problème d'optimisation. Les positions initiales des particules sont généralement choisies au hasard dans l'espace de recherche multidimensionnel.

2. **Évaluation :** Chaque particule est évaluée en fonction d'une fonction de fitness ou d'objectif, qui mesure la qualité de la solution qu'elle représente.

3. **Mise à jour des vitesses :** La vitesse de chaque particule est ajustée en fonction de sa propre meilleure position connue (best personal) et de la meilleure position connue de l'essaim (best global). Cela simule le comportement social où les individus apprennent des succès des autres membres du groupe ainsi que de leurs propres expériences.

4. **Mise à jour des positions :** Les particules se déplacent ensuite dans l'espace de recherche en fonction de leur vitesse mise à jour. Cela peut les amener à explorer de nouvelles régions de l'espace de recherche.

5. **Iteration :** Les étapes d'évaluation et de mise à jour sont répétées jusqu'à ce qu'un critère d'arrêt soit atteint, tel qu'un nombre maximum d'itérations ou un seuil de qualité de solution.

## Quels usages ?

L'algorithme d'optimisation par essaims de particules (PSO) est utilisé dans de nombreux domaines et pour divers types de problèmes d'optimisation concrets.

1. **Ingénierie électrique** : Le PSO est utilisé pour optimiser la conception de circuits électriques, le dimensionnement de transformateurs, et la planification de systèmes de distribution d'électricité. Il est également appliqué dans l'optimisation de la stabilité des systèmes d'énergie électrique et dans la conception de contrôleurs PID (Proportionnel-Intégral-Dérivatif).

2. **Réseaux de neurones** : Dans le domaine de l'intelligence artificielle, le PSO peut être employé pour optimiser les poids et la structure des réseaux de neurones artificiels.

3. **Robotique** : Les robots peuvent utiliser le PSO pour optimiser leurs trajectoires ou coordonner les mouvements dans un essaim de robots afin d'accomplir des tâches de manière collaborative.

4. **Optimisation de portefeuille** : En finance, le PSO peut aider à trouver le meilleur ensemble d'investissements, c'est-à-dire un portefeuille qui optimise le retour sur investissement tout en minimisant le risque.

5. **Aérodynamique** : En conception de véhicules et en ingénierie aérospatiale, le PSO est utilisé pour optimiser la forme des pièces pour réduire la résistance de l'air ou de l'eau et améliorer les performances globales.

6. **Sélection de caractéristiques** : Dans le traitement de données, le PSO peut être utilisé pour sélectionner les caractéristiques les plus pertinentes pour un modèle prédictif, ce qui est particulièrement utile dans les cas où le nombre de caractéristiques est très élevé.

7. **Optimisation de processus** : Il est également utilisé pour optimiser les processus industriels, comme les paramètres dans les processus chimiques pour maximiser la production ou minimiser les coûts.

8. **Planification d'itinéraires** : Le PSO peut optimiser les itinéraires de transport, par exemple, pour minimiser la distance parcourue par les véhicules de livraison.

9. **Calibrage de modèles** : Il peut être utilisé pour calibrer les paramètres de modèles complexes en science de l'environnement, comme les modèles climatiques ou hydrologiques.

10. **Conception d'antennes** : En génie électrique, le PSO est employé pour optimiser la conception d'antennes pour améliorer leur performance.

Ces applications ne sont qu'un échantillon des nombreuses utilisations possibles du PSO. L'adaptabilité et la simplicité de l'algorithme le rendent attrayant pour résoudre des problèmes d'optimisation dans presque tous les domaines de la recherche et de l'industrie.
___

## Forces

L'algorithme d'optimisation par essaims de particules (PSO) présente plusieurs avantages et points forts qui le rendent attrayant dans divers domaines de recherche et d'application pratique :

1. **Simplicité d'implémentation :** PSO est relativement simple à comprendre et à implémenter par rapport à d'autres algorithmes d'optimisation plus complexes.

2. **Peu de paramètres à ajuster :** PSO ne nécessite l'ajustement que de quelques paramètres clés, comme le nombre de particules, les coefficients d'accélération et le facteur d'inertie. Cela rend l'algorithme moins sensible au surparamétrage.

3. **Flexibilité et Adaptabilité :** PSO peut être adapté pour optimiser des problèmes continus ou discrets et peut gérer des espaces de recherche de grande dimension.

4. **Efficacité sur des problèmes non linéaires :** PSO est efficace pour trouver des solutions dans des cas où la fonction objectif est non linéaire et où il n'y a pas d'information sur le gradient.

5. **Capacité de parallélisation :** Les calculs relatifs à chaque particule sont indépendants les uns des autres, ce qui rend l'essaim naturellement apte à la parallélisation, et donc à un gain de performance sur des architectures parallèles.

6. **Efficacité computationnelle :** PSO peut souvent converger vers une solution acceptable en moins d'itérations que d'autres méthodes d'optimisation globale.

7. **Robustesse :** PSO est relativement robuste face au piégeage dans des optima locaux, surtout dans des versions adaptées de l'algorithme qui incluent des mécanismes pour éviter ce problème.

8. **Haut potentiel d'hybridation :** PSO peut être facilement combiné avec d'autres techniques d'optimisation pour améliorer sa performance ou pour aborder des aspects spécifiques d'un problème.
___

## Inconvénients

Bien que l'Optimisation par Essaims de Particules soit un outil puissant et polyvalent pour de nombreux problèmes d'optimisation, il présente également des inconvénients et des limites qui doivent être pris en compte :

1. **Convergence prématurée :** PSO peut parfois converger trop rapidement vers une solution qui n'est pas optimale (un optimum local), en particulier dans des espaces de recherche complexes avec de nombreux optima locaux.

2. **Dépendance aux paramètres :** La performance de PSO dépend fortement des valeurs des paramètres choisis, comme les coefficients d'accélération et le facteur d'inertie. Choisir les bons paramètres peut être non trivial et peut nécessiter une expérimentation approfondie ou une connaissance du domaine.

3. **Difficultés avec les espaces de recherche de haute dimension :** Bien que PSO puisse gérer des espaces de recherche de grande dimension, sa performance peut se dégrader à mesure que le nombre de dimensions augmente, car l'espace de recherche devient exponentiellement plus grand.

4. **Stagnation :** Les particules peuvent parfois stagner si elles ne trouvent pas de meilleures positions, ce qui amène l'ensemble de l'essaim à manquer de diversité et à s'arrêter prématurément.

5. **Balance exploration-exploitation :** Trouver le bon équilibre entre l'exploration de l'espace de recherche (recherche de nouvelles solutions potentielles) et l'exploitation (amélioration des solutions existantes) est crucial. PSO peut avoir du mal à maintenir cet équilibre tout au long du processus d'optimisation.

6. **Problèmes dynamiques ou en mutation :** Dans les problèmes où le paysage d'optimisation change avec le temps, PSO peut avoir du mal à s'adapter aux nouvelles conditions sans modifications ou extensions spécifiques.

7. **Échelles de valeur et unités :** Comme PSO utilise des distances dans l'espace de recherche, les problèmes où les variables ont des échelles ou des unités très différentes peuvent nécessiter une normalisation ou un prétraitement pour s'assurer que l'essaim n'est pas biaisé en faveur d'une dimension particulière.

Pour atténuer ces inconvénients, des variantes de PSO ont été développées, incluant des stratégies pour maintenir la diversité dans l'essaim, des mécanismes d'adaptation des paramètres, et des versions multi-objectifs pour gérer plusieurs fonctions de fitness simultanément. En outre, la combinaison de PSO avec d'autres méthodes d'optimisation peut aussi contribuer à surmonter certains de ces inconvénients.

## Python demonstration

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction de coût - Sphère
def sphere(x):
    return sum(x**2)

# Définition du problème
d = 10  # Nombre de dimensions (variables de décision)
VarSize = d  # Taille de la matrice des dimensions est simplement le nombre de dimensions
LMin = -10  # Limite inférieure des dimensions
LMax = 10  # Limite supérieure des dimensions

# Paramètres PSO
MaxIt = 1000  # Nombre maximal d'itérations
n = 100  # Taille de la population (taille de l'essaim)
w = 1  # Poids d'inertie
wDamp = 0.99  # Rapport d'amortissement de poids d'inertie
c1 = 1.5  # Coefficient d'apprentissage local
c2 = 2.0  # Coefficient d'apprentissage global
VMax = 0.1 * (LMax - LMin)  # Limites max de la vitesse
VMin = -VMax  # Limites min de la vitesse

# Initialisation
class Particle:
    def __init__(self):
        self.Position = np.random.uniform(LMin, LMax, VarSize)
        self.Velocity = np.zeros(VarSize)
        self.Cost = sphere(self.Position)
        self.Best = {'Position': self.Position.copy(), 'Cost': self.Cost}

particles = [Particle() for _ in range(n)]
GlobalBest = min(particles, key=lambda x: x.Cost).Best

BestCost = []

# Boucle principale PSO
for it in range(MaxIt):
    for particle in particles:
        # Mise à jour de la vitesse
        particle.Velocity = (w * particle.Velocity +
                             c1 * np.random.uniform(size=VarSize) * (particle.Best['Position'] - particle.Position) +
                             c2 * np.random.uniform(size=VarSize) * (GlobalBest['Position'] - particle.Position))

        
        # Appliquer les limites de vitesse
        particle.Velocity = np.maximum(particle.Velocity, VMin)
        particle.Velocity = np.minimum(particle.Velocity, VMax)
        
        # Mise à jour de la position
        particle.Position += particle.Velocity
        
        # Effet miroir de vitesse si on sort des limites
        IsOutside = np.logical_or(particle.Position < LMin, particle.Position > LMax)
        particle.Velocity[IsOutside] = -particle.Velocity[IsOutside]
        
        # Appliquer les limites de position
        particle.Position = np.maximum(particle.Position, LMin)
        particle.Position = np.minimum(particle.Position, LMax)
        
        # Évaluation
        particle.Cost = sphere(particle.Position)
        
        # Mise à jour du meilleur local
        if particle.Cost < particle.Best['Cost']:
            particle.Best['Position'] = particle.Position.copy()
            particle.Best['Cost'] = particle.Cost
            
            # Mise à jour du meilleur global
            if particle.Best['Cost'] < GlobalBest['Cost']:
                GlobalBest = particle.Best.copy()
    
    BestCost.append(GlobalBest['Cost'])
    # print(f"Iteration {it+1}: Best Cost = {BestCost[it]}")
    w *= wDamp

# Affichage des résultats
plt.figure()
plt.semilogy(BestCost, linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Best Cost')
plt.grid(True)
plt.show()

# Afficher la meilleure solution
print("Best Solution:", GlobalBest['Position'])
print("Best Cost:", GlobalBest['Cost'])
```
