# Titre

## Algorithme : Qu'est-ce que c'est ?

L'algorithme "Branch and Bound" (Branche et Limite en français) est une méthode générale d'optimisation pour résoudre des problèmes de décision et d'optimisation combinatoire discrets. Cet algorithme est souvent utilisé pour des problèmes où les solutions sont représentées par des arbres de décision, comme dans les cas des problèmes de voyageur de commerce (TSP), de sac à dos (knapsack), ou de programmation en nombres entiers. 

Voici comment fonctionne l'algorithme Branch and Bound :

1. **Branching (Branche)**: Le processus commence par la division (branching) du problème initial en sous-problèmes plus petits et plus gérables. Cette division est souvent représentée sous forme d'arbre, où chaque nœud représente un sous-ensemble des choix possibles.

2. **Bounding (Limite)**: Pour chaque sous-problème, une borne (limite) est calculée. Cette borne représente une évaluation (souvent optimiste) de la meilleure solution possible dans ce sous-arbre. Si cette borne est moins bonne que la meilleure solution déjà trouvée, le sous-arbre correspondant peut être éliminé (ou "coupé") de la recherche.

3. **Élagage (Pruning)**: L'élagage est un aspect crucial de Branch and Bound. Il permet d'éviter d'explorer des parties de l'arbre de décision qui ne mèneront pas à une meilleure solution. Cela est souvent réalisé en comparant les bornes des sous-problèmes avec la meilleure solution actuelle.

4. **Recherche**: L'algorithme parcourt l'arbre, en appliquant les étapes de Branching et Bounding, pour trouver la meilleure solution possible. La recherche peut être effectuée de manière récursive, en explorant en profondeur d'abord (DFS) ou en largeur d'abord (BFS).

5. **Optimalité**: Lorsque l'algorithme termine, la meilleure solution trouvée est garantie d'être optimale sous réserve que toutes les bornes soient calculées correctement.

Branch and Bound est puissant car il permet de résoudre de manière exacte des problèmes qui sont autrement très difficiles à traiter, comme les problèmes NP-difficiles. Cependant, sa performance dépend fortement de la qualité des stratégies de branchement et d'élagage utilisées.

## Quels usages ?

L'algorithme Branch and Bound est utilisé dans divers domaines pour résoudre des problèmes d'optimisation complexes. Voici quelques-uns de ses usages principaux :

1. **Problème du voyageur de commerce (TSP)**: C'est l'un des usages les plus classiques du Branch and Bound. Le TSP consiste à trouver le chemin le plus court permettant à un voyageur de visiter une série de villes et de revenir à son point de départ. L'algorithme explore différentes permutations de villes pour trouver le chemin optimal.

2. **Problèmes de programmation en nombres entiers**: Ces problèmes impliquent la recherche de solutions entières optimales sous des contraintes linéaires. Le Branch and Bound est souvent utilisé pour résoudre des problèmes de programmation linéaire mixte, où certaines variables sont contraintes à prendre des valeurs entières.

3. **Problème du sac à dos (Knapsack Problem)**: Dans ce problème, l'objectif est de maximiser la valeur totale des objets placés dans un sac à dos, tout en respectant une limite de poids. Le Branch and Bound est utilisé pour explorer différentes combinaisons d'objets.

4. **Planification et ordonnancement**: Le Branch and Bound est utilisé pour résoudre des problèmes de planification, comme l'ordonnancement de tâches dans une usine ou l'allocation de ressources dans un projet, où il faut optimiser certains critères comme le temps ou le coût.

5. **Optimisation de réseaux**: Dans la conception et l'optimisation de réseaux (comme les réseaux de télécommunications ou de distribution), le Branch and Bound peut aider à trouver le chemin optimal ou la configuration de réseau minimisant les coûts ou maximisant les performances.

6. **Bioinformatique**: Dans des domaines comme l'alignement de séquences d'ADN, le Branch and Bound peut être utilisé pour trouver l'alignement optimal entre deux séquences, en tenant compte de diverses contraintes et coûts.

7. **Jeux et stratégies**: Pour certains jeux de stratégie comme les échecs ou le Go, le Branch and Bound peut être utilisé pour analyser les arbres de décision et déterminer les meilleurs coups possibles.

___

## Forces

1. **Généralité**: Il peut être appliqué à une grande variété de problèmes d'optimisation, en particulier ceux qui sont NP-difficiles, comme le problème du voyageur de commerce ou les problèmes de programmation en nombres entiers.

2. **Flexibilité**: Branch and Bound peut être adapté à des problèmes spécifiques en modifiant les stratégies de branchement et de bornage, permettant ainsi une personnalisation selon les besoins du problème.

3. **Réduction de l'espace de recherche**: Grâce à l'élagage efficace des branches non prometteuses, l'algorithme réduit l'espace de recherche, ce qui peut significativement diminuer le temps de calcul pour de nombreux problèmes.
___

## Inconvénients

1. **Complexité exponentielle**: Dans le pire des cas, le nombre de sous-problèmes à explorer peut croître de manière exponentielle avec la taille du problème, ce qui rend l'algorithme impraticable pour de très grands ensembles de données.

2. **Dépendance à la qualité des bornes**: L'efficacité de l'algorithme dépend fortement de la qualité des méthodes de bornage. Si les bornes ne sont pas bien choisies, l'algorithme peut perdre beaucoup de temps à explorer des branches inutiles.

3. **Consommation de mémoire**: Pour des problèmes de grande taille, l'algorithme peut nécessiter une quantité considérable de mémoire pour stocker l'état de l'arbre de recherche et les informations associées.

4. **Performance variable**: La performance de l'algorithme peut varier considérablement en fonction de la nature spécifique du problème. Pour certains problèmes, il peut trouver rapidement une solution optimale, alors que pour d'autres, il peut se révéler moins efficace que d'autres approches heuristiques ou approximatives.

5. **Nécessité de heuristiques pour les grands problèmes**: Pour les problèmes de très grande taille, il peut être nécessaire de combiner Branch and Bound avec des heuristiques ou des algorithmes approximatifs pour obtenir des solutions dans un délai raisonnable.

## Solving with base algorithm

```python
import sys
from queue import PriorityQueue

class Node:
    # A Node in Branch and Bound Tree.
    def __init__(self, path, reduced_matrix, cost, vertex, level):
        self.path = path               # Path traversed until now.
        self.reduced_matrix = reduced_matrix  # Reduced cost matrix.
        self.cost = cost               # Total cost so far.
        self.vertex = vertex           # Current vertex.
        self.level = level             # Level in decision tree.

    # Defines comparison operation for PriorityQueue.
    def __lt__(self, other):
        return self.cost < other.cost

# Function to reduce the cost matrix and calculate its cost.
def reduce_matrix(matrix, n):
    # Reduce each row and calculate row reduction cost.
    row_reduce = [min(row) if min(row) != sys.maxsize else 0 for row in matrix]
    # Reduce each column and calculate column reduction cost.
    col_reduce = [min(matrix[i][j] for i in range(n)) if min(matrix[i][j] for i in range(n)) != sys.maxsize else 0 for j in range(n)]

    # Reduce the matrix elements.
    for i in range(n):
        for j in range(n):
            if matrix[i][j] != sys.maxsize and row_reduce[i] != sys.maxsize and col_reduce[j] != sys.maxsize:
                matrix[i][j] -= (row_reduce[i] + col_reduce[j])

    # Calculate total reduction cost.
    cost = sum(row_reduce) + sum(col_reduce)
    return cost

# Function to create a deep copy of a matrix.
def copy_matrix(matrix, n):
    new_matrix = [row[:] for row in matrix]
    return new_matrix

# TSP using Branch and Bound.
def TSP_Branch_and_Bound(TSP):
    n = len(TSP.cities)
    # Create a distance matrix.
    distances = [[0 if i == j else TSP.get_distance(TSP.cities[i][0], TSP.cities[j][0]) for j in range(n)] for i in range(n)]

    # Priority Queue to store live nodes of the search tree.
    pq = PriorityQueue()
    initial_matrix = copy_matrix(distances, n)
    initial_cost = reduce_matrix(initial_matrix, n)

    # Create the first node and add it to the list.
    initial_node = Node(path=[0], reduced_matrix=initial_matrix, cost=initial_cost, vertex=0, level=1)
    pq.put(initial_node)

    # Initialize minimum cost to infinity.
    min_cost = sys.maxsize
    best_path = []

    # Explore nodes using Branch and Bound.
    while not pq.empty():
        current_node = pq.get()

        # Check if all cities are visited.
        if current_node.level == n:
            current_path = current_node.path + [0]
            current_cost = current_node.cost
            if current_cost < min_cost:
                min_cost = current_cost
                best_path = current_path
            continue

        # Iterate through all vertices to find feasible vertices.
        for i in range(n):
            if distances[current_node.vertex][i] != sys.maxsize and i not in current_node.path:
                new_matrix = copy_matrix(current_node.reduced_matrix, n)

                # Set outgoing edges and the incoming edge of the selected vertex as infinity.
                for j in range(n):
                    new_matrix[current_node.vertex][j] = sys.maxsize
                    new_matrix[j][i] = sys.maxsize
                new_matrix[i][0] = sys.maxsize

                # Calculate the cost of the child node.
                cost = current_node.cost + current_node.reduced_matrix[current_node.vertex][i] + reduce_matrix(new_matrix, n)
                new_node = Node(path=current_node.path + [i], reduced_matrix=new_matrix, cost=cost, vertex=i, level=current_node.level + 1)
                pq.put(new_node)

    best_path.reverse()
    # Return the minimum cost and the best path found.
    return best_path, min_cost

# Example usage
path, cost = TSP_Branch_and_Bound(TSP)
print("Minimum cost:", cost)
TSP.draw_path(path)
```

## Solving using PulP library

```python
import pulp
from itertools import product

def solve_tsp_with_pulp(tsp):
    n = len(tsp.cities)
    distances = [[0 if i == j else tsp.get_distance(tsp.cities[i][0], tsp.cities[j][0]) for j in range(n)] for i in range(n)]

    # Create the problem variable to contain the problem data
    prob = pulp.LpProblem("TSP", pulp.LpMinimize)

    # A matrix of binary variables. x[i][j] is 1 if the path goes from city i to j.
    x = pulp.LpVariable.dicts("x", (range(n), range(n)), 0, 1, pulp.LpBinary)

    # Objective function - minimize the total distance
    prob += pulp.lpSum(distances[i][j] * x[i][j] for i in range(n) for j in range(n))

    # Constraints
    for i in range(n):
        prob += pulp.lpSum(x[i][j] for j in range(n) if i != j) == 1 # Exactly one outgoing edge
        prob += pulp.lpSum(x[j][i] for j in range(n) if i != j) == 1 # Exactly one incoming edge

    # Subtour elimination constraints
    u = pulp.LpVariable.dicts("u", range(n), 0, n-1, pulp.LpContinuous)
    for i, j in product(range(1, n), repeat=2):
        if i != j:
            prob += u[i] - u[j] + n * x[i][j] <= n-1

    # Solve the problem
    prob.solve(pulp.apis.PULP_CBC_CMD(msg=False))

    # Extract the solution
    tour = []
    for i in range(n):
        for j in range(n):
            if pulp.value(x[i][j]) == 1:
                tour.append((i, j))

    # Return the ordered tour and the total cost
    return tour, pulp.value(prob.objective)

def reconstruct_path(tour, n):
    path = [0] # Path init with the first city
    while len(path) < n:
        for i, j in tour:
            if i == path[-1]:
                path.append(j)
                break
    path.append(0)    
    return path

# Example usage
tour, cost = solve_tsp_with_pulp(TSP)
print("Tour: ", tour)
ordered_path = reconstruct_path(tour, len(TSP.cities))

print("Minimum cost:", cost)
print("Path:", ordered_path)

# Draw the path using TSP's draw_path method
TSP.draw_path(ordered_path)
```

## Où utiliser Branch and Bound ?

1. [**PuLP (Python)**](https://coin-or.github.io/pulp/) : PuLP est une bibliothèque Python pour la programmation linéaire. Elle permet aux utilisateurs de formuler leurs problèmes d'optimisation de manière très lisible et intuitive, et peut être utilisée avec différents solveurs, y compris ceux qui implémentent le branch and bound. PuLP se distingue par sa facilité d'utilisation et son intégration dans l'écosystème Python, ce qui en fait un choix idéal pour les projets nécessitant une optimisation linéaire et entière en Python.

2. [**CPLEX (C++, Java, Python, entre autres)**](https://www.ibm.com/docs/en/icos/12.9.0?topic=cplex-callable-library-c-api-reference-manual) : CPLEX est un puissant solveur d'optimisation développé par IBM. Il est largement utilisé dans l'industrie et la recherche académique pour résoudre des problèmes d'optimisation linéaire, entière et quadratique. CPLEX dispose d'une interface pour plusieurs langages, notamment C++, Java et Python.

3. [**SCIP (C)**](https://scipopt.org) : SCIP est une bibliothèque de résolution de problèmes d'optimisation combinatoire, qui inclut des algorithmes de branchement et de limitation. Elle est écrite en C et est particulièrement réputée pour sa flexibilité et son efficacité dans la résolution de problèmes d'optimisation entière.

4. [**COIN-OR (C++)**](https://www.coin-or.org) : Le Computational Infrastructure for Operations Research (COIN-OR) est une initiative pour développer des logiciels open-source pour l'analyse et la résolution de problèmes en recherche opérationnelle. Plusieurs de ses projets, écrits en C++, implémentent l'algorithme Branch and Bound.

5. [**MATLAB Optimization Toolbox (MATLAB)**](https://www.mathworks.com/products/matlab.html) : La boîte à outils d'optimisation de MATLAB propose des fonctions pour résoudre des problèmes d'optimisation linéaire, quadratique, non linéaire et d'optimisation entière mixte. Bien que MATLAB ne soit pas open-source, il est largement utilisé dans les milieux académiques et industriels.

6. [**GLPK (GNU Linear Programming Kit) (C)**](https://www.gnu.org/software/glpk/) : GLPK fournit un solveur pour les problèmes de programmation linéaire (LP) et de programmation entière linéaire (ILP). Il est écrit en C et est une option open-source populaire pour l'optimisation linéaire et entière.
