# Newton-Raphson Method and Quasi-Newton methods

# Understanding The Algorithms

## Newton-Raphson method

Newton-Raphson method, a root-finding algorithm, is used to find successively better approximations to the roots (or zeroes) of a real-valued function. It starts with a function `f`, its derivative `f′`, and an initial guess `x0` for a root of `f`.

**Process**:
1. **Initial Guess**: Start with an initial guess `x0`.
2. **Iteration**: Repeat the formula `xn+1 = xn - f(xn)/f'(xn)` until reaching sufficient precision.
3. **Geometric Interpretation**: The x-intercept of the tangent of the graph of `f` at `xn` provides the next approximation `xn+1`.
4. **Convergence**: If the function satisfies certain conditions and the initial guess is close, it converges at least quadratically near the zero.

**Challenges**:
- Requires direct calculation of the derivative, which may not be easily obtainable.
- Convergence issues if the root has multiplicity greater than one or if roots are close together【12†source】【13†source】【14†source】【15†source】【16†source】.

## Quasi-Newton methods

Quasi-Newton methods are used for finding zeroes or local maxima and minima of functions, providing an alternative to Newton's method. They are particularly useful when the Jacobian or Hessian is unavailable or too expensive to compute at every iteration.

**Key Features**:
1. **Hessian Matrix Approximation**: Instead of calculating the Hessian matrix, an estimate `B` is used, avoiding time-consuming calculations.
2. **Quadratic Model Building**: A quadratic model of the objective function is built at each iteration, using the approximated Hessian matrix.
3. **Search Direction**: Determines the direction to move in the function space, guided by the approximate Hessian.
4. **Update Rule**: The Hessian approximation is updated based on gradient information from successive iterations, using methods like the Sherman-Morrison formula.

**Popular Algorithms**:
- **DFP (Davidon-Fletcher-Powell)**: Focuses on estimating the inverse of the Hessian matrix.
- **BFGS (Broyden-Fletcher-Goldfarb-Shanno)**: Estimates the Hessian matrix itself.
- **L-BFGS**: A low-memory extension of BFGS, suitable for large-scale problems.

## Commonalities and differences

### Commonalities
- **Goal**: Both aim to find roots or extrema of functions.
- **Foundation**: Based on the principle of using iterative methods to approximate solutions.
- **Convergence**: Both methods improve the approximation in successive iterations.

### Differences
- **Derivative Requirement**: Newton-Raphson requires direct calculation of derivatives, while Quasi-Newton methods use approximations.
- **Complexity**: Newton-Raphson's direct use of second derivatives can be more computationally intensive, especially for complex functions.
- **Practicality**: Quasi-Newton methods are often more practical for complex optimization problems due to their balance between efficiency and ease of implementation.
- **Hessian Matrix**: Direct computation and inversion of the Hessian matrix are needed in Newton-Raphson, whereas Quasi-Newton methods work with approximations or estimates of it.
- **Application Scope**: Quasi-Newton methods are more versatile and widely used in various fields, including large-scale optimization problems and deep learning, thanks to their efficiency in handling non-differentiable functions and large datasets.

## Usage Examples

### Newton-Raphson method examples

1. **Finding Roots of Polynomials**: For a cubic polynomial $x^3 - 2x + 1$, starting with an initial guess of $x_0 = -1.5$, the method iteratively approximates the root to a desired precision.

2. **Solving Exponential Equations**: To solve $ln(x) = 2$, an initial guess near the root is chosen, and the method iteratively approximates the solution to three decimal places.

3. **Non-Linear Equations**: For the equation $x^x = 5$, starting with $x_0 = 2$, the Newton-Raphson method finds an accurate solution after a few iterations.

4. **Trigonometric Equations**: Solving $cos(x) = x^2$ involves finding the zeros of $f(x) = cos(x) - x^2$. Initial guesses close to the expected roots are used to find the solutions.

5. **Combining Exponential and Polynomial Functions**: For $e^x = 2 - x^2$, the method identifies two solutions by starting with initial guesses near the intercept points of $e^x$ and $2-x^2$.

### Quasi-Newton method examples

1. **Training Deep Neural Networks (DNNs)**: Quasi-Newton methods, particularly Kronecker-factored block-diagonal BFGS and L-BFGS, are used for training DNNs. These methods offer efficient training processes with high-dimensional data, and in tests with autoencoder models, they demonstrated superior or comparable performance to first-order stochastic methods and KFAC.

2. **Solving Stochastic Optimization Problems**: Stochastic L-BFGS methods are applied to strongly convex problems, using sampled Hessian-vector products. These methods have shown linear convergence and are effective in noisy gradient scenarios common in machine learning.

3. **Natural Gradient Methods**: Generalized Gauss-Newton and natural gradient methods, employing quasi-Newton approaches, have been developed for deep learning. These methods use block-diagonal approximations to the Fisher information matrix and are effective in various optimization scenarios.

4. **Implementation in DNN Training**: Kronecker-factored quasi-Newton methods, including K-BFGS and K-BFGS(L), use layer-wise second-order information and are efficient in terms of memory and computational resources. They have been successfully applied to training multi-layer feed-forward neural network models, showing potential for extension to convolutional and recurrent neural networks.

## Strengths

### Newton-Raphson method strengths

1. **Fast Convergence**: It is one of the fastest methods for finding roots, converging rapidly towards the solution.

2. **Quadratic Convergence Rate**: The method has a quadratic convergence rate (rate of convergence is 2), meaning the number of accurate digits roughly doubles with each iteration.

3. **Accuracy**: It is highly effective in obtaining precise results for a root, especially useful when refining solutions initially approximated by other methods.

4. **Extension to Multiple Dimensions**: The Newton-Raphson method can be easily adapted for multidimensional problems, making it versatile and broadly applicable.

### Quasi-Newton method strengths

1. **No Need for Hessian Matrix Inversion**: A significant advantage of Quasi-Newton methods is the elimination of the need to invert the Hessian matrix. In Newton's method, inverting the Hessian is typically required and can be computationally expensive. Quasi-Newton methods, in contrast, generate an estimate of the inverse Hessian directly.

2. **Versatility in Finding Zeroes or Extrema**: Quasi-Newton methods are adaptable for finding zeroes or local maxima and minima of functions. They are particularly useful when the Jacobian or Hessian is either unavailable or too costly to compute in each iteration.

3. **Applicability in Complex Problems**: These methods are suitable for complex optimization problems, especially where the objective function is not twice-differentiable, making the Hessian matrix unavailable or expensive to calculate.

4. **Efficiency in Large-scale Problems**: Quasi-Newton methods are more efficient than Newton's method in large-scale optimization problems due to their less demanding computational requirements.

## Weaknesses

### Weaknesses of the Newton-Raphson method

1. **Requirement of Derivative**: The method requires the calculation of the first (and sometimes second) derivative of the function, which can be complex or even unfeasible for certain functions.

2. **Initial Guess Sensitivity**: The convergence of the method is highly dependent on the initial guess; a poor guess can lead to divergence or convergence to an unintended root.

3. **Poor Performance on Multiple Roots**: If a function has multiple roots, especially close together, the method may struggle to converge correctly.

4. **Non-Convergence for Certain Functions**: For functions with horizontal tangents (where the derivative is zero), the method can fail to converge.

### Weaknesses of Quasi-Newton methods

1. **Approximation of Hessian**: Quasi-Newton methods rely on approximating the Hessian matrix, which can lead to less accurate solutions compared to methods that compute the Hessian directly.

2. **Convergence Rate**: While more efficient in some scenarios, the convergence rate of Quasi-Newton methods is generally slower than the Newton-Raphson method, especially near the solution.

3. **Complexity in Implementation**: The methods can be more complex to implement, particularly in choosing and updating the Hessian approximation.

4. **Sensitivity to Starting Point**: Similar to Newton-Raphson, Quasi-Newton methods can be sensitive to the choice of the starting point, affecting their convergence and effectiveness.

## Python Demonstration

### Root determination with Newton-Raphson method

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize a random number generator
rng_engine = np.random.default_rng(seed=42)

def newton_method(f, df, x0, tol, max_iter):
    """
    Implement Newton's method for finding a root of a real-valued function.

    Args:
    - f: The function for which the root is to be found.
    - df: The derivative of the function f.
    - x0: The initial guess for the root.
    - tol: The tolerance for the absolute value of f to decide convergence.
    - max_iter: The maximum number of iterations to perform.

    Returns:
    - xn: The estimated root of the function. Returns None if a solution is not found within the given tolerance or iterations.
    """
    xn = x0  # Current estimate for the root

    for n in range(max_iter):
        fxn = f(xn)  # Evaluate function at current estimate
        
        # Check if the current estimate is close enough to be considered a root
        if abs(fxn) < tol:
            return xn
        
        dfxn = df(xn)  # Evaluate derivative at current estimate
        
        # Check for zero derivative (slope), which would prevent further progress
        if dfxn == 0:
            print('Zero derivative. No solution found.')
            return None

        # Update the estimate using Newton's method
        xn = xn - fxn/dfxn

    # If maximum iterations are reached without convergence
    print('Exceeded maximum number of iterations. No solution.')
    return None


f = lambda x: x**2 - 9
# f:
# The function for which we are trying to find the roots. In this case, it's a quadratic function f(x) = x^2 - 9.

df = lambda x: 2*x
# df:
# The derivative of the function f. This is used in the Newton-Raphson method to update the current guess towards
# the root. For the given function f(x) = x^2 - 9, the derivative is df(x) = 2x.

# Parameters for Newton-Raphson method
nb_guess = 10
# nb_guess:
# The number of initial guesses (x0) to try in the Newton-Raphson method. Using multiple guesses can help find
# different roots of the function, especially for functions with multiple roots.

tol = 0.000001
# tol:
# The tolerance level for determining when a root has been found. The Newton-Raphson iteration will stop when the
# absolute value of the function at the current guess is less than this tolerance. A smaller tolerance leads to a
# more accurate root but may require more iterations.

max_iter = 10
# max_iter:
# The maximum number of iterations allowed for the Newton-Raphson method. This parameter limits the number of times the
# algorithm will update the guess in search of the root. It prevents the method from running indefinitely in cases where
# it may not converge.

# The found solutions storage
roots = []

for _ in range(nb_guess):
    # Generate a random initial guess within the range [-10, 10]
    x0 = rng_engine.uniform(low=-10, high=10)
    
    # Use Newton's method to find a root starting from x0
    root = newton_method(f, df, x0, tol, max_iter)
    
    # If a root is found, round it and check if it's unique
    if root is not None:
        root = round(root, 5)
        if root not in roots:
            roots.append(root)

# Plotting
x_values = np.linspace(-10, 10, 400)  # Create a range of x-values
y_values = f(x_values)  # Compute y-values for each x

plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label='f(x) = x^2 - 9')  # Plot the function
plt.axhline(0, color='black',linewidth=0.5)  # Add x-axis
plt.axvline(0, color='black',linewidth=0.5)  # Add y-axis

# Plotting each root
for root in roots:
    plt.scatter(root, f(root), color='red', zorder=5)  # Mark the root
    plt.annotate(f'Root ≈ {root:.6f}', (root, f(root)), textcoords="offset points", xytext=(10,10), ha='center', color='red')

plt.title('Newton-Raphson Method')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

<div class="alert alert-block alert-warning">
This version of the Newton-Raphson method stops after finding one root. Since the demonstration function has two roots, the algorithm has been "tuned" to run multiple times with different starting points (x0), allowing it to find all the roots.
</div>

### Demonstration of L-BFGS algorithm on Himmelblau's function

#### Key Components of the demonstration

Following Python code demonstrates the use of the L-BFGS (Limited-memory Broyden–Fletcher–Goldfarb–Shanno) optimization algorithm applied to Himmelblau's function. The goal is to find the minima of this function starting from multiple random initial points.

1. **Himmelblau Function Definition:**
   - Himmelblau's function is defined as: <br>
   $ f(x, y) = (x^2 + y - 11)^2 + (x + y^2 - 7)^2 $ <br>
   where, $ x $ and $ y $ are the variables corresponding to `x[0]` and `x[1]` in the bellow Python code, respectively.
   
   - It is a well-known test function for optimization algorithms, having multiple local minima.

2. **Gradient of the Himmelblau Function:**
   - The gradient is calculated to guide the optimization process.

3. **L-BFGS Algorithm Implementation:**
   - An implementation of the L-BFGS algorithm, specifically adapted for the 2D case of Himmelblau's function.
   
   - The algorithm uses a backtracking line search with the Armijo rule for step size determination.

4. **Finding the Nearest Known Solution:**
   - After finding a solution, the script calculates the nearest known solution from a predefined set of known solutions of Himmelblau's function.

5. **Visualization:**
   - The script visualizes the optimization paths using a contour plot of Himmelblau's function.


6. **Multiple Random Starts:**
   - The optimization is performed multiple times, each starting from a different random initial point within a specified range.

7. **Result Analysis:**
   - A table is generated, comparing the found solutions with the nearest known solutions and displaying the distances.

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize a random number generator without a fixed seed for varied results
rng_engine = np.random.default_rng(seed=None)

def himmelblau_function(x):
    """
    Compute the Himmelblau function, a multi-modal function used in optimization.

    Args:
    - x: A NumPy array or a list with two elements representing the input coordinates.

    Returns:
    - The value of the Himmelblau function at the given coordinates.
    """
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

def grad_himmelblau_function(x):
    """
    Calculate the gradient of the Himmelblau function.

    Args:
    - x: A NumPy array or a list with two elements representing the input coordinates.

    Returns:
    - A NumPy array containing the gradient of the Himmelblau function at the given coordinates.
    """
    df_dx = 4 * x[0] * (x[0]**2 + x[1] - 11) + 2 * (x[0] + x[1]**2 - 7)
    df_dy = 2 * (x[0]**2 + x[1] - 11) + 4 * x[1] * (x[0] + x[1]**2 - 7)
    return np.array([df_dx, df_dy])

def l_bfgs(f, grad_f, initial_guess, max_iter=100, memory=10):
    """
    Implement the L-BFGS optimization algorithm for 2D cases.

    Args:
    - f: The function to minimize.
    - grad_f: The gradient of the function.
    - initial_guess: Initial guess for the solution.
    - max_iter: Maximum number of iterations.
    - memory: Number of previous iterations to remember for approximation.

    Returns:
    - x: The optimized solution.
    - x_path: The path taken to reach the solution.
    """
    x = initial_guess
    Hk = np.eye(2)  # Initial approximation of the inverse Hessian

    # Lists to store past 'm' s and y values
    s_list, y_list = [], []
    x_path = [x.copy()]

    for _ in range(max_iter):
        grad = grad_f(x)
        p = -Hk @ grad  # Determine search direction

        # Line search with Armijo rule
        step_size, armijo_factor, c = 1, 0.5, 0.0001
        while f(x + step_size * p) > f(x) + c * step_size * np.dot(grad, p):
            step_size *= armijo_factor

        x_new = x + step_size * p
        s = x_new - x
        y = grad_f(x_new) - grad

        # Check for convergence
        if np.linalg.norm(s) < 1e-5 or np.linalg.norm(y) < 1e-5:
            break

        # Update lists
        if len(s_list) >= memory:
            s_list.pop(0)
            y_list.pop(0)
        s_list.append(s)
        y_list.append(y)

        # Update Hk
        rho = 1.0 / (y @ s)
        Hk = (np.eye(2) - rho * np.outer(s, y)) @ Hk @ (np.eye(2) - rho * np.outer(y, s)) + rho * np.outer(s, s)

        x = x_new
        x_path.append(x.copy())

    return x, np.array(x_path)

def nearest_known_solution(point, known_solutions):
    """
    Find the nearest known solution to a given point.

    Args:
    - point: The point to compare against known solutions.
    - known_solutions: Array of known solutions.

    Returns:
    - The nearest known solution to the provided point.
    """
    distances = np.linalg.norm(known_solutions - point, axis=1)
    nearest_index = np.argmin(distances)
    return known_solutions[nearest_index]

# Running L-BFGS from different random starts
found_solutions = []
nearest_known_solutions = []

# Number of random starts
num_starts = 3
x0 = np.array([0, 0])

# Known solutions of the Himmelblau function
known_solutions = np.array([[3, 2], [-2.805118, 3.131312], [-3.779310, -3.283186], [3.584428, -1.848126]])

# Create a grid for the contour plot
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = himmelblau_function([X, Y])

# Different colors for different paths
colors = ['red', 'blue', 'green', 'purple', 'orange']

fig, axs = plt.subplots(2, 1, figsize=(10, 8))  # Increased height for additional space for legend

# Upper plot for the function
axs[0].contourf(X, Y, Z, levels=50, cmap='viridis')

for i, point in enumerate(known_solutions):
    axs[0].plot(point[0], point[1], 'yx', markersize=20, label='Known Solutions' if i == 0 else "")

# Resetting the found solutions and nearest known solutions lists
found_solutions = []
nearest_known_solutions = []
table_data = []

# Optimization runs from random starting points
for i in range(num_starts):
    # Generate a random starting point
    x0 = rng_engine.uniform(-5, 5, 2)
    # Perform L-BFGS optimization
    solution, solution_path = l_bfgs(himmelblau_function, grad_himmelblau_function, x0, max_iter=2000)
    found_solutions.append(solution)
    nearest_known_solutions.append(nearest_known_solution(solution, known_solutions))

    # Calculate distances to nearest known solutions
    distance_x = np.linalg.norm(found_solutions[i][0] - nearest_known_solutions[i][0])
    distance_y = np.linalg.norm(found_solutions[i][1] - nearest_known_solutions[i][1])

    row = [f"Path {i+1}", found_solutions[i], nearest_known_solutions[i], [round(distance_x, 5), round(distance_y, 5)]]
    table_data.append(row)

    # Plot optimization paths
    axs[0].plot(solution_path[:, 0], solution_path[:, 1], 'o-', color=colors[i], markersize=3, linewidth=1, label=f'Path {i+1}' if i == 0 else "")
    axs[0].plot(solution_path[0, 0], solution_path[0, 1], 'o', color=colors[i], markersize=8)
    axs[0].plot(solution_path[-1, 0], solution_path[-1, 1], 'x', color=colors[i], alpha=0.8, markersize=13)

axs[0].set_title(f'Himmelblau function optimization with {num_starts} starts')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# Lower plot for the table
axs[1].axis('off')  # Turn off the axis for the table
table = axs[1].table(cellText=table_data, colLabels=["Path", "Found solution", "Nearest known solution", "Distance"], loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 3)  # Adjusted scale for reduced margins

# Add custom legend
legend_elements = [plt.Line2D([0], [0], color='y', lw=4, label='Known solutions', marker='x', markersize=10)]
for i, color in enumerate(colors[:num_starts]):
    legend_elements.append(plt.Line2D([0], [0], color=color, lw=2, label=f'Path {i+1}', marker='o', markersize=8))

fig.legend(handles=legend_elements, loc='lower center', bbox_to_anchor=(0.5, 0.4), ncol=num_starts + 1)


plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout for better visibility
plt.show()
```

## Practical Optimization Tools

1. [**NumPy (Python):**](https://numpy.org/) NumPy, a fundamental package for scientific computing in Python, is extensively used for numerical operations. NumPy is integral to the Python data science stack and is often used in conjunction with SciPy and other libraries to solve optimization problems.

2. [**Julia (Julia Language):**](https://julialang.org) Julia, a high-performance programming language for technical computing, features robust optimization packages. One such package is `Optim.jl`, which offers a variety of optimization algorithms, including Newton and quasi-Newton methods. Its strength lies in the language's speed and ease of use, making it a popular choice for tasks requiring high computational efficiency.

3. [**GNU Octave:**](https://octave.org) GNU Octave, primarily intended for numerical computations, provides a convenient interface for solving optimization problems. It includes functions for applying Newton and quasi-Newton methods in its optimization toolbox. Octave is widely recognized for its MATLAB-like syntax, making it a go-to option for users familiar with MATLAB seeking an open-source alternative.

The `stats` package includes functions for optimization, and `optimx` enhances these capabilities, providing an array of optimization methods, including Newton and quasi-Newton algorithms. R's popularity in the statistical and data analysis community makes these packages widely used for optimization problems in these domains.

5. [**Eigen (C++):**](https://eigen.tuxfamily.org/index.php?title=Main_Page) Eigen, a C++ template library for linear algebra, includes capabilities for solving optimization problems. While primarily focused on matrix operations, its versatility allows for the implementation of Newton and quasi-Newton methods within a high-performance, low-level computing environment. Eigen is well-regarded for its efficiency and is widely used in applications that require heavy linear algebra computations.
